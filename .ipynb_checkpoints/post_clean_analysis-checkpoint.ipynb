```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Read Data

```python
# Read data in.

filename = 'adhoc_data/monthly_expenses_cleaned.csv'
monthly = pd.read_csv(filename)
monthly = monthly.drop('Unnamed: 0', axis=1)
```

```python
# Get salary, location, and age columns. Filter out spouse salaries.

# Need to convert salaries to numeric type.
salaries = monthly[(monthly.owner == 'diarist') & (monthly.title == 'salary')][['diary_index', 'value']]
salaries.value = pd.to_numeric(salaries.value)
salaries = salaries.groupby('diary_index').value.sum()

locations = monthly[monthly.title == 'location'].groupby('diary_index').value.first()

ages = monthly[monthly.title == 'age'].groupby('diary_index').value.first()
```

```python
# Pandas preprocessing.

salaries = pd.DataFrame(salaries).rename_axis({'value': 'salary'}, axis=1)
locations = pd.DataFrame(locations).rename_axis({'value': 'location'}, axis=1)
ages = pd.DataFrame(ages).rename_axis({'value': 'age'}, axis=1)
```

```python
# Merge into `headlines` DataFrame.

_ = salaries.merge(locations, left_index=True, right_index=True)
headlines = _.merge(ages, left_index=True, right_index=True)
```

```python
def which_state(location):
    '''Return the parsed state code for a given location.'''
    txt = location.split(',')[-1].strip()
    if txt == txt.upper():
        return txt
    else:
        return np.nan
```

## Simple Linear Regression

`ln(salary) ~ b0 + b1(age)`

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# Fit SLR model.  Drop null and zero-valued salaries.  Drop nonnumeric ages.

def is_numeric(x):
    '''Return NaN if nonnumeric.'''
    try:
        x = int(x)
    except ValueError:
        x = np.nan
    return x
```

```python
_df = headlines[(headlines.salary > 0) & (headlines.age.apply(is_numeric).notnull())]

y = np.log(_df.salary).values.reshape(-1, 1)
X = _df.age.astype(int).values.reshape(-1, 1)

lr.fit(X, y)
```

```python
# Predict values.

fitline_x = np.linspace(np.min(X), np.max(X))
fitline_y = lr.predict(fitline_x.reshape(-1,1))
```

```python
# Plot.  Jitter age around random unform range.

j = .35
jitter = np.random.uniform(-j, j, X.shape[0]).reshape(-1,1)

plt.figure(figsize=(16,6))
plt.title('Scatterplot of age on log-salary', fontsize=14)
plt.xlabel('age')
plt.ylabel('log-salary')
plt.scatter(X + jitter, y, alpha=0.6)
plt.plot(fitline_x, fitline_y, color='red');
```

```python
# Score the model.

lr.score(X, y)
```

## Ranking of monthly expenses vs. salary

What do peoples' monthly spending habits tell us about thie salary and vice versa?

### Reshaping

- Don't drop spouse/significant other information.
- Numeric information: groupby `diary_index` and `title`, and take the sum.
- Object information: groupby `diary_index` and `title`, and **proof that results are unique**.
- Concat the results and pivot so that each row is a single `diary_index`.

```python
monthly.head()
```

```python
# All columns other than diary_index are Object type, so split
# into two dfs by row, groupby and sum the numeric one, and
# join back together.

monthly = monthly.dropna(subset=['title'])
monthly.title = monthly.title.apply(lambda x: '_'.join(x.split()))
num_titles = ['salary', 'monthly_paycheck', 'rent', 'student_loan',
             'utilities', 'health_insurance', 'charity', 'gym',
             'entertainment', 'savings', 'mortgage', 'auto', 'income',
             'childcare', 'credit', 'misc', 'auto']
object_titles = ['occupation', 'industry', 'location', 'age']
```

```python
# ugh. Rogue nonparseable value.

ix_to_fix = monthly[monthly.value == '348..40'].index
monthly.loc[ix_to_fix, 'value'] = '348.40'
```

```python
num = monthly[monthly.title.isin(num_titles)]
obj = monthly[monthly.title.isin(object_titles)]
```

```python
num .value = num.value.astype(float)
num = num.groupby(['diary_index', 'title']).sum()
```

```python
# Print the number of dupes in each column: where counts are greater than 1.

test_uniqueness = obj.groupby(['diary_index', 'title']).count()
for c in test_uniqueness.columns:
    num_dupes = test_uniqueness[test_uniqueness[c] > 1].shape[0]
    print("Number of duplicates in column {0:<12}{1:d}".format(c.upper(), num_dupes))
```

It's safe to take the first row of group!

```python
# Group `obj`, take first row, and drop columns so that it has the same cols as `num`.

obj = obj.groupby(['diary_index', 'title']).first()
cols_to_drop = [c for c in obj.columns if c not in num.columns]
obj = obj.drop(cols_to_drop, axis=1)
```

```python
# Concat one atop the other, reset index.

long = pd.concat([num, obj], axis=0)
long = long.reset_index()
```

```python
# There are a few nulls.  Safe to fill with 0's because they are all
# in numeric-type fields.

n = long[long.value.isnull()].shape[0]
u = long[long.value.isnull()].title.unique()
print("Number of null values: {0:d}".format(n))
print("Fields where null:")
for u_ in u:
    print("  -{0:s}".format(u_))
    
long.value = long.value.fillna(0)
```

```python
wide = long.pivot(index='diary_index', columns='title', values='value')
```

```python
wide[num_titles] = wide[num_titles].fillna(0.0)
wide[object_titles] = wide[object_titles].fillna(np.nan)
```

### Ready to do stuff with `wide`

Plot log-salary on log-rent.

```python
nulls_dropped = wide[(wide.salary > 0) & (wide.rent > 0)]  # Zero vals are very annoying.
```

```python
X = np.log(nulls_dropped.salary)
y = np.log(nulls_dropped.rent)
```

```python
lr_log = LinearRegression()
lr_log.fit(X.reshape(-1, 1), y)
```

```python
plt.figure(figsize=(16,4))
plt.title('Scatterplot of log-salary on log-rent', fontsize=14)
plt.xlabel('log-salary')
plt.ylabel('log-rent')

fitline_x = np.linspace(np.min(X), np.max(X), num=1000)
fitline_y = lr_log.predict(fitline_x.reshape(-1, 1))

plt.scatter(X, y, alpha=0.6);
plt.plot(fitline_x, fitline_y, color='red');
```

Find the biggest residuals...who are those heathens that spend way too much on rent? Who are they and are there any commonalities between them?

```python
y_hat = lr_log.predict(X.reshape(-1, 1))
resid = y - y_hat
```

```python
# Plot with line at the mean and lines at 2-sigma's.

plt.figure(figsize=(16,4))
plt.title('Scatterplot of residuals (log-salary regressed on log-rent)', fontsize=14)
plt.xlabel('(index)')
plt.ylabel('residual')

std = np.std(resid)

plt.scatter(np.arange(0, resid.shape[0]), resid, alpha=0.6)
plt.plot(np.arange(0, resid.shape[0]), np.zeros(resid.shape[0]), color='black', linestyle='--')
plt.plot(np.arange(0, resid.shape[0]), np.full(resid.shape[0], 2*std), color='grey', linestyle='--')
plt.plot(np.arange(0, resid.shape[0]), np.full(resid.shape[0], -2*std), color='grey', linestyle='--');
```

```python
# Build new sorted DF, with overspenders at the top.

resid = pd.Series(resid)
resid.name = 'rent_resid'
_ = pd.concat([nulls_dropped, resid], axis=1)
rent_order = _.sort_values(by='rent_resid', ascending=False)
rent_order['rent_rank'] = rent_order.rent_resid.rank(ascending=False)
```

```python
shrinkcols = ['salary', 'rent', 'location', 'age', 'industry', 'occupation', 'rent_resid', 'rent_rank']
rent_order[shrinkcols].head(15)
```

### Function for iterating over columns

```python
from sklearn.model_selection import cross_val_score
```

```python
expense_columns = ['auto', 'charity', 'childcare', 'credit', 'entertainment', 'gym',
                  'health_insurance', 'misc', 'mortgage', 'rent', 'savings', 'student_loan',
                  'utilities']
```

```python
def func(x_name, y_names):
    '''
    Function to execute the same analysis as above, looping through expense columns.
    Return a map from expense colname to a df of the above analysis (`ordered`).
    Plot along the way.
    '''
    df_dict = {}
    for y_name in y_names:

        nulls_dropped = wide[(wide[x_name] > 0) & (wide[y_name] > 0)]

        X = nulls_dropped[x_name]
        y = nulls_dropped[y_name]

        X_log = np.log(nulls_dropped[x_name])
        y_log = np.log(nulls_dropped[y_name])

        lr = LinearRegression()

        untransformed_score = cross_val_score(lr, X_log.values.reshape(-1, 1), y, cv=5).mean()
        logged_score = cross_val_score(lr, y_log.values.reshape(-1, 1), y, cv=5).mean()

        plot_score = untransformed_score
        logged = ""
        if logged_score > untransformed_score:
            X = X_log
            y = y_log
            logged = " (log model)"
            plot_score = logged_score

        lr.fit(X.values.reshape(-1, 1), y)

        y_hat = lr.predict(X.values.reshape(-1, 1))
        resid = y - y_hat

        print(y_name.upper())
        print("="*len(y_name))
        print("Beta: {0:.3f}".format(lr.coef_[0]))
        print("R2:   {0:.3f}\n".format(plot_score))
        
        # Scatterplot
        plt.figure(figsize=(16,4))
        plt.title('Scatterplot of {0:s} on {1:s}{2:s}'.format(x_name, y_name, logged), fontsize=14)
        plt.xlabel('{0:s}'.format(x_name))
        plt.ylabel('{0:s}'.format(y_name))

        fitline_x = np.linspace(np.min(X), np.max(X), num=1000)
        fitline_y = lr.predict(fitline_x.reshape(-1, 1))

        plt.scatter(X, y, alpha=0.6);
        plt.plot(fitline_x, fitline_y, color='red');


        # Residplot
        plt.figure(figsize=(16,4))
        plt.title('Scatterplot of residuals ({0:s} regressed on {1:s}{2:s})'.format(x_name, y_name, logged), fontsize=14)
        plt.xlabel('(index)')
        plt.ylabel('residual')

        std = np.std(resid)

        plt.scatter(np.arange(0, resid.shape[0]), resid, alpha=0.6)
        plt.plot(np.arange(0, resid.shape[0]), np.zeros(resid.shape[0]), color='black', linestyle='--')
        plt.plot(np.arange(0, resid.shape[0]), np.full(resid.shape[0], 2*std), color='grey', linestyle='--')
        plt.plot(np.arange(0, resid.shape[0]), np.full(resid.shape[0], -2*std), color='grey', linestyle='--');

        resid = pd.Series(resid)
        residname = y_name + '_resid'
        resid.name = residname
        _ = pd.concat([nulls_dropped, resid], axis=1)
        ordered = _.sort_values(by=residname, ascending=False)
        rankname = y_name + '_rank'
        ordered[rankname] = ordered[residname].rank(ascending=False)

        df_dict[y_name] = ordered
    return df_dict
```

```python
d = func('salary', expense_columns)
```

## KMeans with Daily Text

```python
dt_path = 'data_reshape/daily_text.csv'
dailytext = pd.read_csv(dt_path)
```

```python
dailytext.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
```

```python
# Takes a long time to run!

vec = CountVectorizer()
sparse = vec.fit_transform(dailytext.day_rawtext)

ks = range(2,11)
inertias = []
clst_counts = []

for k in ks:
    clst = KMeans(n_clusters=k)
    predicted_cluster_centers = clst.fit_predict(sparse)
    clst_count = np.unique(predicted_cluster_centers, return_counts=True)
    clst_counts.append(clst_count)
    inertias.append(clst.inertia_)
```

```python
# Elbow plot of KMeans results.
# 4 looks like a decent value of K to me.

plt.plot(ks, inertias);
```
